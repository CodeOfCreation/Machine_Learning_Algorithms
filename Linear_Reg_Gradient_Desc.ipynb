```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
m=29.19
```

```python
class GDreg:
    
    def __init__(self,lr=0.01,epochs=1000):
        self.m=29.19
        self.b=-120
        self.lr=lr
        self.epochs=epochs

    def fit(self,x,y):
        for i in range(self.epochs):
            loss_slope=-2* np.sum(y-self.m*x.ravel()-self.b)
            self.b=self.b-self.lr*loss_slope
            
        print(loss_slope,self.b)
        print(self.b)
```

```python
GD = GDreg(0.001,100)
GD.fit(x,y)
```
